### Small sanity check

```python
data = {
    'Имя': ['Алексей', 'Мария', 'Иван', 'Елена'],
    'Роль': ['Python Dev', 'Data Analyst', 'DevOps', 'Product Manager'],
    'Зарплата_руб': [250000, 180000, 220000, 300000],
    'Кофе_чашек_в_день': [4, 2, 5, 1]
}

df = pd.DataFrame(data)

print("График сотрудников и зарплат:")
print(df)
```

```python
result = df["Зарплата_руб"].agg(['mean', 'median', 'min', 'max'])
print("\nСтатистика по зарплатам:")
print(result)
```

```python
plt =  mtpl.pyplot

data = {'Год': [2010, 2011, 2012, 2013, 2014],
        'Продажи': [200, 300, 400, 500, 600]}

df = pd.DataFrame(data)

df.plot(x='Год', y='Продажи', kind='line')

plt.title('Динамика продаж по годам')
plt.xlabel('Год')
plt.ylabel('Продажи')
plt.show()
```
